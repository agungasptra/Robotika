```python
# Install TensorFlow
!pip install tensorflow

# Import libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load dataset (contoh: dataset iris)
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data
y = iris.target

# Normalisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Pisahkan data menjadi train dan test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Buat model neural network sederhana
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(3, activation='softmax')  # Output layer dengan 3 kelas (sesuai dengan jumlah kelas pada dataset iris)
])

# Compile model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Training model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Evaluasi model pada data test
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')
```
